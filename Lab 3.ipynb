## Loops Through All Files In Directory And Finds The Counts Of Gender

```python
files = glob.glob('DischargeNotes-09/*')
stringToMatch = 'Sex:'
matchedLine = ''
maleCount = 0
femaleCount = 0
noGenderCount = 0

for file in files:
    with open(file, "r") as f:
        for line in f:
            if stringToMatch in line:
                matchedLine = line
                gender = matchedLine.split("Sex:", 1)[1]
                if 'M' in gender:
                    maleCount += 1
                elif 'F' in gender:
                    femaleCount += 1
                else:
                    noGenderCount += 1
                break
                
                
print("Male: " + str(maleCount))
print("Female: " + str(femaleCount))
print("Not listed: " + str(noGenderCount))


dictionary = {'gender':['male', 'female', 'not listed'], 'count':[maleCount, femaleCount, noGenderCount]}

df = pd.DataFrame(dictionary)
```
